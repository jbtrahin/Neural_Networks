```python
# Import our dependencies
import pandas as pd
import numpy as np
import os
import matplotlib as plt
from pathlib import Path
import sklearn as skl
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Import our input dataset
file_path = Path('../Challenge/Resources/charity_data.csv')
charity_df = pd.read_csv(file_path, index_col="EIN")
charity_df.head()
```

```python
# Store names in a dataframe
charity_name_df = pd.DataFrame(data=charity_df['NAME'], index=charity_df.index)
charity_name_df.head()
```

```python
# Drop name column
charity_df_drop = charity_df.copy()
charity_df_drop.drop(columns=['NAME'], inplace=True)

charity_df_drop.head()
```

```python
# Generate our categorical variable list
charity_cat = charity_df_drop.dtypes[charity_df.dtypes == "object"].index.tolist()
charity_cat

# Check the number of unique values in each column
charity_df_drop[charity_cat].nunique()
```

```python
# TAKING CARE OF APPLICATION_TYPE BINNING
# Check the unique value counts to see if binning is needed
application_type_count = charity_df_drop.APPLICATION_TYPE.value_counts()
application_type_count
```

```python
# Let's assume cut_off_value=20000 for categorization in "others" as there is such a big difference between T3 and other application types.
# Find valus to replace
replace_application_type = list(application_type_count[application_type_count<20000].index)

# Replace in DataFrame
for application_type in replace_application_type:
    charity_df_drop.APPLICATION_TYPE = charity_df_drop.APPLICATION_TYPE.replace(application_type, "others")

# Check binning was successful
charity_df_drop.APPLICATION_TYPE.value_counts()
```

```python
# TAKING CARE OF CLASSIFICATION BINNING
classification_count = charity_df_drop.CLASSIFICATION.value_counts()
pd.set_option('display.max_rows', 500)
classification_count
```

```python
# Let's assume cut_off_value=10000 for categorization in "others" as there is such a big difference between C1000 and other classification.
# Find valus to replace
replace_classification = list(classification_count[classification_count<10000].index)

# Replace in DataFrame
for classification in replace_classification:
    charity_df_drop.CLASSIFICATION = charity_df_drop.CLASSIFICATION.replace(classification, "others")

# Check binning was successful
charity_df_drop.CLASSIFICATION.value_counts()
```

```python
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(charity_df_drop[charity_cat]), index=charity_df_drop.index)

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(charity_cat)
encode_df.head()
```

```python
# Drop Special_Considerations_N as it is a binary variable and it would be redundant data.
encode_df = encode_df.drop("SPECIAL_CONSIDERATIONS_N",1)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop originals
charity_df_final = charity_df_drop.merge(encode_df, left_index=True, right_index=True)
charity_df_final = charity_df_final.drop(charity_cat,1)
charity_df_final.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = charity_df_final["IS_SUCCESSFUL"].values
X = charity_df_final.drop("IS_SUCCESSFUL",1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
number_input_features = len(X_train_scaled[0])
number_input_features
```

```python
# Define the model - deep neural net
# A good rule of thumb for a basic neural network is to have two to three times the amount of neurons in the hidden layer as the number of inputs.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 32
hidden_nodes_layer2 = 16
hidden_nodes_layer3 = 16

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model (let's start with 100 epochs)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)

# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Our target is to achieve a target predictive accuracy higher than 75%.
# To otimize we'll increase the numbers of neurons by hidden layer to reach close to three times the number of imputs (31 inputs x 3 = 92).

# Define a new model

number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 48
hidden_nodes_layer2 = 24
hidden_nodes_layer3 = 12

optimized_nn = tf.keras.models.Sequential()

# First hidden layer
optimized_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
optimized_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
optimized_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
optimized_nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
optimized_nn.summary()
```
